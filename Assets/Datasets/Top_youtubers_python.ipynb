```python
# I used Pthon function to call Data and import data from youtube API using request
def get_channel_stats(youtube, channel_id):
    request = youtube.channels().list(
        part='snippet,statistics',id=channel_id
    )
    response = request.execute()
    if 'items' in response:
        data = {
            "channel_name":response['items'][0]['snippet']['title'],
            "total_subscribers":response['items'][0]['statistics']['subscriberCount'],
            "total_views":response['items'][0]['statistics']['viewCount'],
            "total_videos":response['items'][0]['statistics']['videoCount'],
        }
        return data
    else:
        return None
# i call out my function with the argument from sample of my Dataset Before
channel_id = "UC_aEa8K-EOJ3D6gOs7HcyNg"
get_channel_stats(youtube, channel_id)
```

```python
# i brought up my Table which has no data for my Analysis 
df1 = pd.read_csv(r"C:\Users\my\Downloads\youtube_data_united-kingdom.csv")
df1.head(3)
```

```python
# From here i extracted channel_ids from my Table and used the split to separate channel_name and channel-id
channel_ids = df1['NOMBRE'].str.split('@').str[-1].unique()
print(channel_ids)
```

```python
#After i had all the channel_ids i created Array where i will put all my Data
all_channel_data = []
#i made a loop from channel_ids i extracted from Nombre
for channel_id in channel_ids:
    channel_data = get_channel_stats(youtube, channel_id)
    if channel_data:
        all_channel_data.append(channel_data)

df = pd.DataFrame(all_channel_data)
print(df)
```

```python
#i concatenated my new Table with data i received from the API with the Previous table
combined_data = pd.concat([df1,df], axis=1)
combined_data
```

```python
#i dropped channel_name from the new table since the Existing table had NOMBRE which had channel_name and channel_id
combined_data.drop('channel_name', axis=1, inplace=True)
```
